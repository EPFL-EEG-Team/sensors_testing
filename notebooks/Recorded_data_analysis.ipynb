```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("../src/eeg_22-04-21-03-46-41.csv", header=None)
#data = pd.read_csv("../src/eeg_22-04-21-03-18-38.csv", header=None)

plt.plot(data[1])
```

```python
length = len(data[1])
fftfreq = np.fft.fftfreq(length,d=2/1000);
amp=np.abs(np.fft.fft(data[1][0:length]))**2
pw = amp[1:length//16]/np.sum(amp[1:length//16])
plt.plot(fftfreq[1:length//16],pw)
```

```python
#https://ieeexplore.ieee.org/stamp/stamp.jsp?tp=&arnumber=8875677
# theta, alpha, beta_low
freq_ranges = [(4, 7), (8, 12), (13, 20)]
summary = [np.mean(pw[np.where((fftfreq>=x)&(fftfreq<=y))]) for (x, y) in freq_ranges]
freq_names = ["theta", "alpha", "beta(low)"]
plt.bar(freq_names, summary)
```
